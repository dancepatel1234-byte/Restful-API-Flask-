Q1.What is a RESTful API?

A RESTful API is an architectural style for an application programming interface (API) that uses HTTP requests to access and manipulate data over the internet. It allows different software systems—like a mobile application and a cloud server—to securely exchange information using a standardized, predictable set of rules.

# Core Components of a Request
When a client application communicates with a RESTful API, the interaction relies on four primary building blocks:   
*  **Endpoint:** The specific URL representing the digital resource you want to access    
*  **HTTP Method:** The verb or command indicating exactly what action you want to perform.   
*  **Headers:** Metadata containing authorization keys, connection types, and the required data format (such as JSON or XML).   
*  **Body:** The actual data payload sent to the server, typically utilized during creation or update actions.

#Key Guiding Principles
To be classified as truly "RESTful," an API must adhere to six specific architectural constraints:

*  **Statelessness:** The server retains no session history about the client. Every incoming request must contain all the data necessary to process it independently.

*  **Client-Server Architecture:** The user interface (client) and data storage (server) operate as completely independent systems, allowing either side to scale or upgrade without affecting the other.

*  **Uniform Interface:** All API requests must use identical communication channels. This requires consistent resource URLs, standardized HTTP methods, and uniform response formats.

*  **Cacheability:** Server responses must declare whether they can be cached by the client. This limits redundant network requests, speeding up overall system performance.

*  **Layered System:** The client cannot tell if it is connected straight to the backend server or interacting through hidden layers like load balancers, proxies, or security firewalls.Standard

**Flask is popular for building APIs due to several key reasons:**

**Lightweight and Minimalist:**

Its microframework nature means it has a small codebase and minimal overhead, making it efficient for building lightweight APIs that focus solely on data exchange.

**Flexibility and Extensibility:**

**Active Community and Ecosystem:**

Flask benefits from a large and active community, providing extensive documentation, tutorials, and a rich ecosystem of extensions and libraries specifically designed for enhancing API functionality.

Q3.  What are HTTP methods used in RESTful API?

RESTful APIs leverage standard HTTP methods (also known as HTTP verbs) to perform actions on resources, aligning with the principles of creating, reading, updating, and deleting (CRUD) data.
The most commonly used HTTP methods in RESTful APIs are:

**GET:**    
Used to retrieve data from the server. It is a safe and idempotent method, meaning it does not alter the server's state and repeated requests yield the same result.

**POST:**     
Used to create new resources or submit data to be processed. It is neither safe nor idempotent as it typically results in the creation of new resources with each request.  

**PUT:**     
Used to update an existing resource or create a new one if it does not exist at a specific URI. It is an idempotent method, as repeated requests with the same data will result in the same state.

**PATCH:**      
Used to apply partial modifications to an existing resource. Unlike PUT, it only sends the changes to be applied, not the entire resource representation. It is generally not considered idempotent without careful implementation.

**DELETE:**    
Used to remove a resource from the server at a specified URI. It is an idempotent method.
Other less frequently used, but still relevant, HTTP methods in RESTful contexts include:

**HEAD:**      
Similar to GET, but retrieves only the header information of a resource without the response body.

**OPTIONS:**      
Used to describe the communication options for the target resource, such as the allowed HTTP methods

Q4. What is the purpose of the @app.route() decorator in Flask

The app.route() decorator in Flask serves the purpose of defining URL routes and associating them with specific Python functions within a Flask web application.

**Here's a breakdown of its purpose:**

**URL Mapping:**   
It maps a given URL path to a Python function that will handle requests made to that URL. When a user accesses a particular URL in their browser, Flask uses this mapping to determine which function should be executed to generate the response.

**Request Handling:**     
The function decorated with @app.route() becomes the "view function" for that particular route. This function contains the logic to process the incoming request, interact with data (e.g., from a database), and finally, return a response (e.g., an HTML page, JSON data, or a redirect).

**Readability and Conciseness:**     
Using a decorator like @app.route() provides a clean and readable way to define routes directly above the function they are associated with, making the code more organized and easier to understand compared to manually adding URL rules.

**Dynamic Routes:**     
It allows for the creation of dynamic routes using variable parts in the URL (e.g., /<username>). These variable parts are then passed as arguments to the decorated view function, enabling the creation of more flexible and interactive web applications.


Q5. What is the role of Flask-SQLAlchemy?


Flask-SQLAlchemy is a Flask extension that makes using SQLAlchemy with Flask easier, providing you tools and methods to interact with your database in your Flask applications through SQLAlchemy.
# Core Roles of Flask-SQLAlchem

1.  **Object-Relational Mapping (ORM):** It acts as a bridge between Python and SQL databases. Instead of writing raw SQL queries, you define tables as Python classes, allowing you to manage database records using standard object-oriented programming.
2. **Session Management:** It automatically handles the opening, closing, and teardown of database sessions per web request. This ensures that your Create, Read, Update, and Delete (CRUD) transactions are safely committed or rolled back to maintain data integrity.
3. **Flask Integration:** By wrapping around standard SQLAlchemy, it abstracts away tedious setup processes. It provides the db instance and helpers like db.Column and db.relationship that integrate perfectly with the Flask application context.



Q6. How do you create a basic Flask application?

To create a basic Flask application, you need to install Flask, write a short Python script to initialize the application and define a route, and then run the development server

Step 1: Install Flask \
Open your terminal or command prompt and install Flask using

    pip:bashpip install flask

Step 2: Create the Application File

Create a new Python file in your project directory and name it app.py. Open the file in a text editor and add the following five lines of code:

    from flask import Flask
    app = Flask(__name__)

    @app.route("/")
    def hello_world():
    return "Hello, World!"

# Code Explanation    
* **from flask import Flask:** Imports the Flask class constructor from the package.    
* **app = Flask(__name__):** Creates an instance of the web application. Flask uses the __name__ argument to know where to look for resources like templates and static assets.    
* **@app.route("/"):** A decorator that binds a URL path to a specific Python function. In this case, the root URL (/) triggers the function below it.    
*  **return "Hello, World!":** The string text that Flask will render directly in the user's web browser.

    def get_data():

Q8 How do you handle POST requests in Flask?

Handling POST requests in Flask involves defining routes that specifically accept the POST HTTP method and then accessing the data sent in the request.   
**Here's how to handle POST requests in Flask:**

-  **Import request:** The request object from Flask is essential for accessing incoming request data.

    from flask import Flask, request, render_template, redirect, url_for

-  **Define a route with methods=['POST']:** When defining your route using the @app.route() decorator, specify that it accepts POST requests using the methods argument. If the route also needs to handle GET requests (e.g., for displaying a form), you can include both.


    @app.route('/submit_form', methods=['GET', 'POST'])
    def submit_form():
        # ...
-  **Check for POST request:** Inside your route function, use request.method == 'POST' to determine if the current request is a POST request. This allows you to differentiate between initial page loads (GET) and form submissions (POST).

    @app.route('/submit_form', methods=['GET', 'POST'])
    def submit_form():
        if request.method == 'POST':
            # Handle POST data
            pass
        else:
            # Handle GET request (e.g., render a form)
            pass

**Access POST data:**


-  **Form data:** If the data is sent from an HTML form with enctype="application/x-www-form-urlencoded" (the default), access it using request.form. This is a dictionary-like object where keys are the name attributes of your form fields.
Python

        username = request.form['username']
        password = request.form['password']
-  **JSON data:** If the data is sent as JSON (e.g., from an API client), use request.get_json(). This parses the JSON body into a Python dictionary.
Python

        data = request.get_json()
        item_name = data['name']
        item_price = data['price']

**Process the data:**

Once you have accessed the data, you can perform necessary operations, such as saving it to a database, performing calculations, or validating inputs.

**Return a response:**
-  After processing the POST request, return an appropriate response, which could be:
-  A redirect to another page using redirect(url_for('some_route')). This is common after form submissions to prevent resubmission on refresh.
-  A rendered template using render_template().
-  JSON data if you are building an API.

To structure a Flask app using Blueprints, you must separate your application into self-contained modules (like authentication, billing, or profile management) and register them to a central application instance using the Application Factory pattern. This prevents circular dependencies and scales cleanly as your project grows.

**1. Recommended Directory Structure**    
The most scalable approach is the functional/divisional hybrid structure, where each Blueprint lives in its own package directory alongside its own routes, templates, and static files.

my_flask_app/    
│
├── app/   
│   ├── __init__.py          # Contains the Application   Factory       
│   ├── extensions.py        # Database and plugin initializations      
│   │
│   ├── auth/                # Authentication Blueprint Package       
│   │   ├── __init__.py      # Defines and exports the Blueprint instance   
│   │   ├── routes.py        # Views and URL endpoints      for auth
│   │   └── templates/       # Blueprint-specific HTML     files
│   │       └── auth/    
│   │           └── login.html    
│   │
│   └── main/                # Main/Core Blueprint Package
│       ├── __init__.py
│       ├── routes.py    
│       └── templates/    
│           └── main/     
│               └── index.html   
│
├── config.py                # Environment configurations   
└── run.py                   # Application entry point     


**2. Define the Blueprint**

Inside your module's __init__.py file, initialize the Blueprint object.python# app/auth/__init__.py

    from flask import Blueprint

    1. Define the blueprint name
    2. Pass __name__ to help locate asset folders
    3. Add an optional URL prefix so all routes start with /auth
    auth_bp = Blueprint('auth', __name__, url_prefix='/auth', template_folder='templates')

    # Import routes at the bottom to avoid circular import errors
    from app.auth import routes

 Instead of using @app.route(), use your newly created blueprint instance decorator (@auth_bp.route).

    # Renders app/auth/templates/auth/login.html
